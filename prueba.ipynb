```python
#hacer tres frames de manera horizontal 
from customtkinter import *  
ventanita=CTk() 
ventanita.geometry("950x500")
ventanita.resizable(False, False)
ventanita.title("ejemplo de frames")

frame1=CTkFrame(ventanita, width=500, height=00, corner_radius=0)
frame1.grid(row=0, column=0, sticky="ns", padx=10, pady=10)
frame1.configure(bg_color="red", fg_color="black", border_width=0)


frame2=CTkFrame(ventanita, width=300, height=400, corner_radius=0)
frame2.grid(row=0, column=1, sticky="ns", padx=10, pady=10)
frame2.configure(bg_color="green", fg_color="black", border_width=0)

frame3=CTkFrame(ventanita, width=300, height=400, corner_radius=0)
frame3.grid(row=0, column=2, sticky="ns", padx=10, pady=10)
frame3.configure(bg_color="blue", fg_color="black", border_width=0)




#en el primer frame pedir los datos de un estudiante (nombre y su calificacion) con sus respectivas etiquetas y entradas
label_nombre=CTkLabel(frame1, text="Nombre:")
label_nombre.grid(row=0, column=0, padx=10, pady=10)
entry_nombre=CTkEntry(frame1)
entry_nombre.grid(row=0, column=1, padx=10, pady=10)

label_calificacion=CTkLabel(frame1, text="Calificacion:")
label_calificacion.grid(row=3, column=0, padx=10, pady=10)
entry_calificacion=CTkEntry(frame1) 
entry_calificacion.grid(row=3, column=1, padx=10, pady=10)


lista_alumnos_aprobados=[]
lista_alumnos_reprobados=[]

def guardar_datos():
    nombre=entry_nombre.get()
    calificacion=entry_calificacion.get()
    if nombre and calificacion:
        try:
            calificacion = float(calificacion)
            if 0 <= calificacion <= 100:
                alumno={"nombre": nombre, "calificacion": calificacion}
                lista_alumnos_aprobados.append(alumno)
                lista_alumnos_reprobados.append(alumno)

                entry_nombre.delete(0, END)
                entry_calificacion.delete(0, END)
            else:
                print("La calificación debe estar entre 0 y 100.")
        except ValueError:
            print("Por favor, ingresa una calificación válida.")
    

def mostrar_aprobados_reprobados():
    for alumno in lista_alumnos_aprobados:
        if alumno['calificacion'] >= 70:
            estado = "Aprobado"
            label_resultado=CTkLabel(frame2, text=f"Nombre: {alumno['nombre']}\nCalificacion: {alumno['calificacion']}\nEstado: {estado}")
            label_resultado.pack(padx=10, pady=10)
        else:
            estado = "Reprobado"
            label_resultado=CTkLabel(frame3, text=f"Nombre: {alumno['nombre']}\nCalificacion: {alumno['calificacion']}\nEstado: {estado}")
            label_resultado.pack(padx=10, pady=10)




boton_guardar=CTkButton(frame1, text="Guardar", command=guardar_datos)
boton_guardar.grid(row=4, column=0, columnspan=2, padx=10, pady=10)

boton_mostrar=CTkButton(frame1, text="Mostrar datos", command=mostrar_aprobados_reprobados)
boton_mostrar.grid(row=5, column=0, columnspan=2, padx=10, pady=10)


#que el boton que aparece en el frame 1 me muestre los aprobados en el frame 2 y los reprobados en el frame 3



ventanita.mainloop()
```
